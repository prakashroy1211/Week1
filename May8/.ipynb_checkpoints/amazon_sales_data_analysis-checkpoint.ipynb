```python
import pandas as pd

# Load the dataset
df = pd.read_csv('amazon.csv')
```

```python
df.head()
```

```python
# Information about dataset
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Product categories and count the products in each category
category_count = df['category'].value_counts()
category_count
```

```python
# Removing unwanted characters from 'rating' column
df['rating'] = df['rating'].str.replace('|', '')
```

```python
df['rating'] = pd.to_numeric(df['rating']) #convert rating to numeric
```

```python
import matplotlib.pyplot as plt
```

```python
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)

# Bar chart
# Show top 10 categories with highest average rating
top_categories = category_rating.head(10)

top_categories.plot(kind='bar', color='green')
plt.title('Top 10 Categories by Average Rating')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
# Convert the column to numeric
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',','') # Remove rupee symbol and commas
df['actual_price'] = pd.to_numeric(df['actual_price'])

prices = df['actual_price']

# Calculate mean, median and standard deviation
mean_price = np.mean(prices)
median_price = np.median(prices)
std_price = np.std(prices)

print(f"Mean Actual Price: {mean_price:.2f}")
print(f"Median Actual Price: {median_price:.2f}")
print(f"Standard Deviation of Actual Price: {std_price:.2f}")
```
